As a first task you need to tokenize first 15 tweets from `file2` by hand. This will allow you to understand the problem from a linguistic point of view. The guidelines for tweet tokenization are as follows:

- Each smiley is a separate token
- Each hashtag is an individual token. Each user reference is an individual token
- If a word has spaces between them then it is converted to a single token
- All punctuations are individual tokens. This includes double-quotes and single quotes also
- A URL is a single token

## Task 2. Implement [Byte-Pair Encoding(BPE)](https://arxiv.org/pdf/1508.07909.pdf) Tokenizer (80 points)

### Task 2.1. Implementation (60 points)

Implement the tokenizer as the BPETokenizer class:
* Implement `train` method that learns merges and builds the vocabulary of the specified `vocab_size` (25 points).
* Implement `tokenize` method that should tokenize the text according to the learnt merges (25 points).

Your code should have docstrings and comments (10 points).

```python
from typing import List
from collections import defaultdict
import re

class BPETokenizer:
    """
    Byte Pair Encoding (BPE) Tokenizer class.

    Args:
        vocab_size (int): The desired size of the vocabulary.

    Attributes:
        vocab_size (int): The desired size of the vocabulary.
        vocab (List[str]): The vocabulary list.
    """
    def __init__(self, vocab_size: int) -> None:
        """
        Initializes the BPETokenizer.

        Args:
            vocab_size (int): The desired size of the vocabulary.
        """
        self.vocab_size = vocab_size
        self.vocab = []

    def _preprocess(self, text: str) -> str:
        """
        Preprocesses the input text by replacing punctuation and extra spaces, and converting spaces to underscores.

        Args:
            text (str): The input text to be preprocessed.

        Returns:
            str: The preprocessed text.
        """
        text = re.sub(r"[,.;@#?!&$]+", " ", text).strip()
        text = re.sub(r"\ +", " ", text)
        text = text.replace(" ", "_")
        text = list(text) + ['_']
        return text

    def _flatten(self, corpus: List[str]) -> List[str]:
        """
        Flattens the input list of sentences into a list of characters after preprocessing.

        Args:
            corpus (List[str]): The input list of sentences.

        Returns:
            List[str]: The flattened list of characters.
        """
        out = []
        for sentence in corpus:
            out.extend(self._preprocess(sentence))
        return out

    def _most_common_pair(self, corpus: List[str]) -> tuple[str, str]:
        """
        Finds the most common pair of consecutive characters in the corpus.

        Args:
            corpus (List[str]): The input list of characters.

        Returns:
            Tuple[str, str]: The most common pair of consecutive characters.
        """
        hashmap = defaultdict(int)
        for i in range(len(corpus) - 1):
            c1, c2 = corpus[i], corpus[i + 1]
            if c1[-1] != '_':
                hashmap[(c1, c2)] += 1
        sorted_items = sorted(hashmap.items(), key=lambda x: x[1], reverse=True)
        if len(sorted_items):
            return sorted_items[:1][0][0]
        return None

    def _update_corpus(self, corpus: List[str], target: tuple[str, str]) -> List[str]:
        """
        Updates the corpus by merging the most common pair in the input target.

        Args:
            corpus (List[str]): The input list of characters.
            target (Tuple[str, str]): The target pair to merge.

        Returns:
            List[str]: The updated corpus.
        """
        out = []
        i = 0
        while i < len(corpus):
            if i + 1 < len(corpus) and (corpus[i], corpus[i + 1]) == target:
                out.append(corpus[i] + corpus[i + 1])
                i += 2
            else:
                out.append(corpus[i])
                i += 1
        return out

    def train(self, corpus: List[str]) -> None:
        """
        Trains the tokenizer on the given corpus using the Byte Pair Encoding algorithm.

        Args:
            corpus (List[str]): The input list of sentences.
        """
        corpus = self._flatten(corpus)
        self.vocab.extend(list(set(corpus)))

        while len(self.vocab) < self.vocab_size and self._most_common_pair(corpus):
            pair = self._most_common_pair(corpus)
            self.vocab.append(pair)
            corpus = self._update_corpus(corpus, target=pair)

    def tokenize(self, text: str) -> List[str]:
        """
        Tokenizes the input text using the trained vocabulary.

        Args:
            text (str): The input text to be tokenized.

        Returns:
            List[str]: The list of tokens.
        """
        text = self._preprocess(text)

        for merged in self.vocab:
            if isinstance(merged, tuple):
                text = self._update_corpus(text, merged)

        out = []
        for token in text:
            if token in self.vocab or len(token) > 1:
                if token[-1] == '_':
                    if len(token[:-1]):
                        out.append(token[:-1])
                else:
                    out.append(token)
            else:
                out.append('<UNK>')

        return out
```

### Task 2.2. Analysis on Tweets Dataset (10 points)

Train the BPE tokenizer on the tweets dataset. Try to tokenize the tweets with the tokenizer of different `vocab_size`. For example, train the BPE tokenizer with `vocab_size` of [base_vocab_size, 250, 500, 750, 1000].
```python
# Your code here
vocab_sizes = [-1, 250, 500, 750, 1000]
mean_lengths = []
tweets = []

for i in range(1, 6):
    with open(f'Assignment1_data/file{i}', 'r') as f:
        for line in f:
            tweets.append(line)

train_size = int(len(tweets) * 0.8)


for vocab_size in vocab_sizes:
    BPE = BPETokenizer(vocab_size)
    BPE.train(tweets[:train_size])
    
    lengths = []
    for sentence in tweets[train_size:]:
        tokenized = BPE.tokenize(sentence)
        lengths.append(len(tokenized))
    
    mean_lengths.append(sum(lengths) / len(lengths))

```
